# Pre Analysis of Climate Sensor Data

##### Imports

```python
# Imports
# Standard
import pandas as pd
import numpy as np
import os

# Plots
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import seaborn as sns
sns.set_theme()
sns.set_palette(sns.color_palette("colorblind"))
sns.set(rc={'figure.facecolor':'white'})

#Plotly
```

##### Load Data

```python
df_climate = pd.read_csv("../../data/raw/Microclimate_Sensor_Readings.csv")
df_sensors = pd.read_csv("../../data/raw/Microclimate_Sensor_Locations.csv")
```

```python
df_climate.head(5)
```

```python
df_sensors.head(5)
```

##### Join datasets

```python
df_climate = df_climate.set_index(['site_id','gateway_hub_id'])
df_sensors = df_sensors.set_index(['site_id','gateway_hub_id'])
df_sensors.columns = [x+'_sensor' for x in df_sensors.columns]
df = df_climate.join(df_sensors,how='inner',rsuffix='_sensor').reset_index()
df.head(5)
```

```python
df.to_csv('../../data/preproccesed/Climate_Data_joined.csv')
```
